# Catalan's Constant $\beta(2)$

Catalan's constant $\beta(2)$ is defined as the infinite sum:

$$\beta(2) = \sum_{k=0}^{\infty} \frac{(-1)^k}{(2k+1)^2} = \frac{1}{1^2} - \frac{1}{3^2} + \frac{1}{5^2} - \frac{1}{7^2} + \dots$$

This is an alternating series. For an alternating series $\sum (-1)^k a_k$ where $a_k > 0$ and $a_k$ decreases monotonically, the error when approximating the sum by the partial sum $S_n$ is less than or equal to the absolute value of the first neglected term, i.e., $|a_{n+1}|$. Therefore, to achieve six-digit precision (error less than $0.5 \times 10^{-6}$), we need the absolute value of the next term to be less than this amount.

Below, we will calculate this sum and mark when six-digit precision is achieved.

```python
import math
import matplotlib.pyplot as plt

def calculate_catalan_constant_beta2(accuracy=1e-6):
    beta_2_sum = 0.0
    k = 0
    accuracy_reached = False
    current_term = 0.0

    partial_sums = []
    error_bounds = []
    accuracy_k_index = -1

    print(f"Calculando \u03B2(2) con una precisi\u00f3n de {accuracy}")
    print("---------------------------------------------------")

    # The actual value of Catalan's constant is approximately 0.91596559
    # We need to ensure the error is less than 0.5 * 10^-6 for 6-digit accuracy.
    # For an alternating series, the error is bounded by the absolute value of the next term.
    error_threshold = 0.5 * accuracy # For 6-digit accuracy, error < 0.5 * 10^-6

    while True:
        denominator = (2 * k + 1)**2
        term = ((-1)**k) / denominator
        beta_2_sum += term
        partial_sums.append(beta_2_sum)

        # Check for accuracy using the next term (as per alternating series error bound)
        if not accuracy_reached:
            next_denominator = (2 * (k + 1) + 1)**2
            next_term_abs_value = 1 / next_denominator
            error_bounds.append(next_term_abs_value)

            if next_term_abs_value < error_threshold:
                accuracy_reached = True
                accuracy_k_index = k
                print(f"T\u00e9rmino {k}: {term:.10f}, Suma parcial: {beta_2_sum:.10f}  <-- \u00a1Precisi\u00f3n de 6 d\u00edgitos alcanzada aqu\u00ed!")
            else:
                print(f"T\u00e9rmino {k}: {term:.10f}, Suma parcial: {beta_2_sum:.10f}")
        else:
            # Continue printing for a few more terms after accuracy is reached to show stability
            if k < 100000 + 10: # Print 10 more terms after reaching accuracy
                next_denominator = (2 * (k + 1) + 1)**2 # Keep calculating error for plot
                next_term_abs_value = 1 / next_denominator
                error_bounds.append(next_term_abs_value)
                print(f"T\u00e9rmino {k}: {term:.10f}, Suma parcial: {beta_2_sum:.10f}")

        if k > 100000 and not accuracy_reached: # Safety break to prevent infinite loops in case accuracy is too strict
            print("\u00a1Advertencia: L\u00edmite de t\u00e9rminos alcanzado sin lograr la precisi\u00f3n deseada!")
            break

        if accuracy_reached and next_term_abs_value < error_threshold * 0.1 and k > accuracy_k_index + 10: # Stop once well within accuracy after a few more terms
             break

        k += 1

    print("---------------------------------------------------")
    print(f"Valor final calculado de \u03B2(2): {beta_2_sum:.10f}")
    return beta_2_sum, partial_sums, error_bounds, accuracy_k_index, error_threshold

# Ejecutar el c\u00e1lculo
calculated_beta2, partial_sums, error_bounds, accuracy_k_index, error_threshold = calculate_catalan_constant_beta2()

# Valor conocido para comparaci\u00f3n (aproximado)
known_beta2 = 0.915965594177219
print(f"Valor conocido de \u03B2(2):     {known_beta2:.10f}")
print(f"Diferencia:                   {abs(calculated_beta2 - known_beta2):.10f}")

# Plotting the convergence
plt.figure(figsize=(12, 6))
add_watermark_to_figure(plt.gcf())
plt.plot(partial_sums, label=r'Suma Parcial $\beta(2)$')
plt.axhline(y=known_beta2, color='r', linestyle='--', label=r'Valor Real de $\beta(2)$ (aprox.)')
plt.axvline(x=accuracy_k_index, color='g', linestyle=':', label=f'Precisi\u00f3n 6 d\u00edgitos alcanzada (k={accuracy_k_index})')
plt.title(r'Convergencia de la Constante de Catalan $\beta(2)$')
plt.xlabel('N\u00famero de T\u00e9rminos (k)')
plt.ylabel('Valor de la Suma')
plt.grid(True)
plt.legend()
plt.xlim(0, accuracy_k_index + 100) # Show a bit more after accuracy is reached
plt.ylim(0.915, 0.917) # Zoom in for better visualization of convergence
plt.show()

plt.figure(figsize=(12, 6))
add_watermark_to_figure(plt.gcf())
plt.plot(error_bounds, label=r'L\u00edmite de Error ($|a_{k+1}|$)')
plt.axhline(y=error_threshold, color='orange', linestyle='--', label=f'Umbral de Precisi\u00f3n (0.5 * 10^-6)')
plt.axvline(x=accuracy_k_index, color='g', linestyle=':', label=f'Precisi\u00f3n 6 d\u00edgitos alcanzada (k={accuracy_k_index})')
plt.title(r'L\u00edmite de Error en la Convergencia de $\beta(2)$')
plt.xlabel('N\u00famero de T\u00e9rminos (k)')
plt.ylabel(r'Valor Absoluto del Siguiente T\u00e9rmino ($|a_{k+1}|$)')
plt.yscale('log') # Use log scale for better visualization of decreasing error
plt.grid(True, which="both", ls="-")
plt.legend()
plt.xlim(0, accuracy_k_index + 100)
plt.show()
```

```python
import math
import matplotlib.pyplot as plt

def calculate_catalan_constant_kummer(accuracy=1e-6):
    kummer_sum = 0.0
    n = 0
    accuracy_reached = False

    # Fixed term from Kummer's acceleration
    fixed_term = (math.pi / 8) * math.log(2 + math.sqrt(3))

    # Initialize tracking variables
    partial_sums_kummer = [fixed_term] # Start with the fixed term
    error_bounds_kummer = []
    accuracy_n_index = -1

    print(f"Calculando \u03B2(2) con la serie de Kummer (precisi\u00f3n de {accuracy})")
    print("-----------------------------------------------------------------------")

    # For 6-digit accuracy, error < 0.5 * 10^-6
    error_threshold = 0.5 * accuracy

    # Add the first term (n=0) of the series part manually to handle the initial binomial coefficient
    term_n0 = (3 / 8) * (1 / (math.comb(0,0) * (2*0 + 1)**2)) # math.comb(0,0) is 1
    kummer_sum = fixed_term + term_n0
    partial_sums_kummer.append(kummer_sum)
    print(f"T\u00e9rmino fijo + serie n=0: {kummer_sum:.10f}")
    n = 1 # Start loop from n=1

    while True:
        # Calculate the current term for the series part
        binomial_coeff = math.comb(2 * n, n)
        term_series_part = (3 / 8) * (1 / (binomial_coeff * (2 * n + 1)**2))

        # Add to the sum
        kummer_sum += term_series_part
        partial_sums_kummer.append(kummer_sum)

        # Check for accuracy using the next term as error bound
        if not accuracy_reached:
            # Calculate the next term for the error bound
            next_n = n + 1
            next_binomial_coeff = math.comb(2 * next_n, next_n)
            next_term_abs_value = (3 / 8) * (1 / (next_binomial_coeff * (2 * next_n + 1)**2))
            error_bounds_kummer.append(next_term_abs_value)

            if next_term_abs_value < error_threshold:
                accuracy_reached = True
                accuracy_n_index = n
                print(f"T\u00e9rmino Kummer n={n}: {term_series_part:.10f}, Suma parcial: {kummer_sum:.10f}  <-- \u00a1Precisi\u00f3n de 6 d\u00edgitos alcanzada aqu\u00ed!")
            else:
                print(f"T\u00e9rmino Kummer n={n}: {term_series_part:.10f}, Suma parcial: {kummer_sum:.10f}")
        else:
            # Continue printing for a few more terms after accuracy is reached to show stability
            if n < accuracy_n_index + 10: # Print a few more terms
                next_n = n + 1
                next_binomial_coeff = math.comb(2 * next_n, next_n)
                next_term_abs_value = (3 / 8) * (1 / (next_binomial_coeff * (2 * next_n + 1)**2))
                error_bounds_kummer.append(next_term_abs_value)
                print(f"T\u00e9rmino Kummer n={n}: {term_series_part:.10f}, Suma parcial: {kummer_sum:.10f}")

        if n > 1000 and not accuracy_reached: # Safety break
            print("\u00a1Advertencia: L\u00edmite de t\u00e9rminos alcanzado sin lograr la precisi\u00f3n deseada con Kummer!")
            break

        if accuracy_reached and n > accuracy_n_index + 10: # Stop once well within accuracy
            break

        n += 1

    print("-----------------------------------------------------------------------")
    print(f"Valor final calculado de \u03B2(2) (Kummer): {kummer_sum:.10f}")
    return kummer_sum, partial_sums_kummer, error_bounds_kummer, accuracy_n_index, error_threshold

# Ejecutar el c\u00e1lculo con la serie de Kummer
calculated_beta2_kummer, partial_sums_kummer, error_bounds_kummer, accuracy_n_index_kummer, error_threshold_kummer = calculate_catalan_constant_kummer()

# Valor conocido para comparaci\u00f3n (aproximado)
known_beta2 = 0.915965594177219
print(f"Valor conocido de \u03B2(2):                         {known_beta2:.10f}")
print(f"Diferencia (Kummer vs Conocido):                 {abs(calculated_beta2_kummer - known_beta2):.10f}")

# Plotting the convergence for Kummer's series
plt.figure(figsize=(12, 6))
add_watermark_to_figure(plt.gcf())
plt.plot(partial_sums_kummer, label=r'Suma Parcial $\beta(2)$ (Kummer)')
plt.axhline(y=known_beta2, color='r', linestyle='--', label=r'Valor Real de $\beta(2)$ (aprox.)')
plt.axvline(x=accuracy_n_index_kummer, color='g', linestyle=':', label=f'Precisi\u00f3n 6 d\u00edgitos alcanzada (n={accuracy_n_index_kummer})')
plt.title(r'Convergencia de la Constante de Catalan $\beta(2)$ (Serie de Kummer)')
plt.xlabel('N\u00famero de T\u00e9rminos (n)')
plt.ylabel('Valor de la Suma')
plt.grid(True)
plt.legend()
plt.xlim(0, accuracy_n_index_kummer + 5) # Show a bit more after accuracy is reached
plt.show()

plt.figure(figsize=(12, 6))
add_watermark_to_figure(plt.gcf())
plt.plot(error_bounds_kummer, label=r'L\u00edmite de Error Kummer ($|a_{n+1}|$)')
plt.axhline(y=error_threshold_kummer, color='orange', linestyle='--', label=f'Umbral de Precisi\u00f3n (0.5 * 10^-6)')
plt.axvline(x=accuracy_n_index_kummer, color='g', linestyle=':', label=f'Precisi\u00f3n 6 d\u00edgitos alcanzada (n={accuracy_n_index_kummer})')
plt.title(r'L\u00edmite de Error en la Convergencia de $\beta(2)$ (Serie de Kummer)')
plt.xlabel('N\u00famero de T\u00e9rminos (n)')
plt.ylabel(r'Valor Absoluto del Siguiente T\u00e9rmino ($|a_{n+1}|$)')
plt.yscale('log') # Use log scale for better visualization of decreasing error
plt.grid(True, which="both", ls="-")
plt.legend()
plt.xlim(0, accuracy_n_index_kummer + 5)
plt.show()
```

## Calculation of Lambda(3) $\lambda(3)$ (using a specific series)

The constant $\lambda(3)$ is often defined as the Dirichlet Lambda function evaluated at 3, which is $\lambda(3) = (7/8)\zeta(3)$, where $\zeta(3)$ is Apéry's constant. Its approximate value is $\mathbf{1.051800040}$.

However, in this notebook, we are asked to calculate the sum of the following specific series, which is presented as $\lambda(3)$:

$$\lambda(3) = 1 + \frac{1}{12} - 4 \sum_{n=1}^{\infty} \frac{1}{(2n+1)^3 (2n-1)(2n+3)}$$

This is a series with a fixed initial part and a summation. For rapid convergence, especially for series of positive terms that decrease monotonically, the absolute value of the next term can serve as a reasonable estimate for the error bound when aiming for a certain precision. We will calculate this sum and mark when six-digit accuracy is achieved (error less than $0.5 \times 10^{-6}$). All output will be in English.

```python
import math
import matplotlib.pyplot as plt

def calculate_lambda3(accuracy=1e-6):
    fixed_part = 1 + (1/12) # Corrected from 1/8 to 1/12
    lambda3_sum = fixed_part
    n = 1 # The sum starts from n=1
    accuracy_reached = False

    partial_sums_lambda3 = [fixed_part]
    error_bounds_lambda3 = []
    accuracy_n_index = -1

    print(f"Calculating \u03BB(3) with accuracy of {accuracy}")
    print("---------------------------------------------------")

    # For 6-digit accuracy, error < 0.5 * 10^-6
    error_threshold = 0.5 * accuracy

    while True:
        denominator_term1 = (2 * n + 1)**3
        denominator_term2 = (2 * n - 1)
        denominator_term3 = (2 * n + 3)

        term = -4 / (denominator_term1 * denominator_term2 * denominator_term3)
        lambda3_sum += term
        partial_sums_lambda3.append(lambda3_sum)

        # Check for accuracy using the absolute value of the next term as error bound
        if not accuracy_reached:
            next_n = n + 1
            next_denominator_term1 = (2 * next_n + 1)**3
            next_denominator_term2 = (2 * next_n - 1)
            next_denominator_term3 = (2 * next_n + 3)
            next_term_abs_value = abs(-4 / (next_denominator_term1 * next_denominator_term2 * next_denominator_term3))
            error_bounds_lambda3.append(next_term_abs_value)

            if next_term_abs_value < error_threshold:
                accuracy_reached = True
                accuracy_n_index = n
                print(f"Term n={n}: {term:.15f}, Partial Sum: {lambda3_sum:.15f}  <-- 6-digit accuracy reached here!")
            else:
                print(f"Term n={n}: {term:.15f}, Partial Sum: {lambda3_sum:.15f}")
        else:
            # Continue printing for a few more terms after accuracy is reached to show stability
            if n < accuracy_n_index + 10: # Print a few more terms
                next_n = n + 1
                next_denominator_term1 = (2 * next_n + 1)**3
                next_denominator_term2 = (2 * next_n - 1)
                next_denominator_term3 = (2 * next_n + 3)
                next_term_abs_value = abs(-4 / (next_denominator_term1 * next_denominator_term2 * next_denominator_term3))
                error_bounds_lambda3.append(next_term_abs_value)
                print(f"Term n={n}: {term:.15f}, Partial Sum: {lambda3_sum:.15f}")

        if n > 1000 and not accuracy_reached: # Safety break
            print("Warning: Term limit reached without achieving desired accuracy for Lambda(3)!")
            break

        if accuracy_reached and n > accuracy_n_index + 10: # Stop once well within accuracy
            break

        n += 1

    print("---------------------------------------------------")
    print(f"Final calculated value of \u03BB(3) (given series): {lambda3_sum:.15f}")
    return lambda3_sum, partial_sums_lambda3, error_bounds_lambda3, accuracy_n_index, error_threshold

# Execute the calculation for Lambda(3)
calculated_lambda3, partial_sums_lambda3, error_bounds_lambda3, accuracy_n_index_lambda3, error_threshold_lambda3 = calculate_lambda3()

# The generally accepted value for Lambda(3) = (7/8)zeta(3)
known_lambda3_dirichlet = 0.875 * 1.2020569031595942 # zeta(3) approx 1.2020569031595942
print(f"Known value of \u03BB(3) (Dirichlet Lambda(3)): {known_lambda3_dirichlet:.15f}")
print(f"Difference (given series vs Dirichlet Lambda(3)): {abs(calculated_lambda3 - known_lambda3_dirichlet):.15f}")

# Plotting the convergence of Lambda(3)
plt.figure(figsize=(12, 6))
add_watermark_to_figure(plt.gcf())
plt.plot(partial_sums_lambda3, label=r'Partial Sum $\lambda(3)$ (given series)')
plt.axhline(y=known_lambda3_dirichlet, color='r', linestyle='--', label=r'Actual Value of $\lambda(3)$ (Dirichlet Lambda(3))')
plt.axvline(x=accuracy_n_index_lambda3, color='g', linestyle=':', label=f'6-digit Accuracy Reached (n={accuracy_n_index_lambda3})')
plt.title(r'Convergence of $\lambda(3)$ (Given Series vs Dirichlet Lambda(3))')
plt.xlabel('Number of Terms (n)')
plt.ylabel('Sum Value')
plt.grid(True)
plt.legend()
plt.xlim(0, accuracy_n_index_lambda3 + 5) # Show a bit more after accuracy is reached
plt.savefig('convergence_lambda3.png') # Save the first plot
plt.show()

plt.figure(figsize=(12, 6))
add_watermark_to_figure(plt.gcf())
plt.plot(error_bounds_lambda3, label=r'Error Bound ($|a_{n+1}|$)')
plt.axhline(y=error_threshold_lambda3, color='orange', linestyle='--', label=f'Accuracy Threshold (0.5 * 10^-6)')
plt.axvline(x=accuracy_n_index_lambda3, color='g', linestyle=':', label=f'6-digit Accuracy Reached (n={accuracy_n_index_lambda3})')
plt.title(r'Error Bound in $\lambda(3)$ Convergence')
plt.xlabel('Number of Terms (n)')
plt.ylabel(r'Absolute Value of Next Term ($|a_{n+1}|$)')
plt.yscale('log') # Use log scale for better visualization of decreasing error
plt.grid(True, which="both", ls="-")
plt.legend()
plt.xlim(0, accuracy_n_index_lambda3 + 5)
plt.savefig('error_bound_lambda3.png') # Save the second plot
plt.show()
```

## Comparison of Convergence for Zeta(3) Series

We will now compare the convergence of three different series expansions for Riemann Zeta function at s=3, often denoted as Apéry's constant, $\zeta(3)$. The known value for $\zeta(3)$ is approximately **1.2020569031595942**.

Here are the three series we will implement and compare:

1.  **Standard Series:**
    $$ \zeta(3) = \sum_{n=1}^{\infty} \frac{1}{n^3} $$

2.  **Series 2 (Accelerated):**
    $$ \zeta(3) = \frac{5}{4} - \sum_{n=2}^{\infty} \frac{1}{n^3 (n^2-1)} $$

3.  **Series 3 (More Accelerated):**
    $$ \zeta(3) = \frac{39}{32} + 4\sum_{n=3}^{\infty} \frac{1}{n^3 (n^2-1)(n^2-4)} $$

We will aim for a 6-digit accuracy (error less than `0.5 * 10^-6`) for each series and observe how many terms are required to achieve this precision, using the absolute value of the next term as an error bound heuristic for comparison. All output will be in English.

```python
import math
import matplotlib.pyplot as plt

# Known value of Zeta(3) (Apéry's constant)
known_zeta3 = 1.2020569031595942
accuracy_threshold = 1e-6
error_bound_target = 0.5 * accuracy_threshold

def calculate_zeta3_standard(error_threshold):
    partial_sums = []
    error_bounds = []
    current_sum = 0.0
    n = 1
    accuracy_n_index = -1
    accuracy_reached = False

    while True:
        term = 1 / (n**3)
        current_sum += term
        partial_sums.append(current_sum)

        # For positive series, the error bound is not simply the next term.
        # However, for comparison of convergence speed, we can use it as a heuristic.
        # A more accurate error bound for sum(1/n^s) is approximately integral(1/x^s dx) from n to inf
        # which for s=3 is 1/(2n^2). But for consistency in comparing 'next term' we'll stick to it.
        next_term_abs_value = 1 / ((n + 1)**3)
        error_bounds.append(next_term_abs_value)

        if not accuracy_reached and next_term_abs_value < error_threshold:
            accuracy_reached = True
            accuracy_n_index = n

        # Safety break for very slow converging series, and to capture a few terms after accuracy
        if n > 50000 and (not accuracy_reached or n > accuracy_n_index + 10):
            if not accuracy_reached:
                print("Warning: Term limit reached without achieving desired accuracy for Standard Zeta(3)!")
            break

        n += 1

    print(f"Standard Zeta(3) calculated. Terms: {n-1}, Final Sum: {current_sum:.15f}")
    print(f"Standard Zeta(3) Accuracy reached at n={accuracy_n_index}")
    return current_sum, partial_sums, error_bounds, accuracy_n_index

def calculate_zeta3_series2(error_threshold):
    partial_sums = []
    error_bounds = []
    fixed_part = 5/4
    current_sum = fixed_part
    # Add the fixed part to partial sums to make plots consistent
    partial_sums.append(current_sum)

    n = 2  # Summation starts from n=2
    accuracy_n_index = -1
    accuracy_reached = False

    while True:
        # Term is negative
        term_value = 1 / (n**3 * (n**2 - 1))
        term = -term_value
        current_sum += term
        partial_sums.append(current_sum)

        next_n = n + 1
        next_term_abs_value = 1 / (next_n**3 * (next_n**2 - 1))
        error_bounds.append(next_term_abs_value)

        if not accuracy_reached and next_term_abs_value < error_threshold:
            accuracy_reached = True
            accuracy_n_index = n

        # Stop after a few terms past accuracy for plot clarity
        if n > 1000 and (not accuracy_reached or n > accuracy_n_index + 10):
            if not accuracy_reached:
                print("Warning: Term limit reached without achieving desired accuracy for Series 2 Zeta(3)!")
            break

        n += 1

    print(f"Series 2 Zeta(3) calculated. Terms: {n-1}, Final Sum: {current_sum:.15f}")
    print(f"Series 2 Zeta(3) Accuracy reached at n={accuracy_n_index}")
    return current_sum, partial_sums, error_bounds, accuracy_n_index

def calculate_zeta3_series3(error_threshold):
    partial_sums = []
    error_bounds = []
    fixed_part = 115/96 # Corrected from 39/32
    current_sum = fixed_part
    # Add the fixed part to partial sums to make plots consistent
    partial_sums.append(current_sum)

    n = 3  # Summation starts from n=3
    accuracy_n_index = -1
    accuracy_reached = False

    while True:
        term_value = 4 / (n**3 * (n**2 - 1) * (n**2 - 4))
        term = term_value # Original series 3 has +4
        current_sum += term
        partial_sums.append(current_sum)

        next_n = n + 1
        # Need to ensure n^2 - 4 is not zero for next_n, which it won't be if next_n >= 3
        next_term_abs_value = 4 / (next_n**3 * (next_n**2 - 1) * (next_n**2 - 4))
        error_bounds.append(next_term_abs_value)

        if not accuracy_reached and next_term_abs_value < error_threshold:
            accuracy_reached = True
            accuracy_n_index = n

        # Stop after a few terms past accuracy for plot clarity
        if n > 1000 and (not accuracy_reached or n > accuracy_n_index + 10):
            if not accuracy_reached:
                print("Warning: Term limit reached without achieving desired accuracy for Series 3 Zeta(3)!")
            break

        n += 1

    print(f"Series 3 Zeta(3) calculated. Terms: {n-1}, Final Sum: {current_sum:.15f}")
    print(f"Series 3 Zeta(3) Accuracy reached at n={accuracy_n_index}")
    return current_sum, partial_sums, error_bounds, accuracy_n_index


print(f"Calculating Zeta(3) series with an accuracy threshold of {accuracy_threshold} (error < {error_bound_target})")
print("--------------------------------------------------------------------------------")

# Execute calculations for all three series
calculated_zeta3_standard, ps_standard, eb_standard, acc_idx_standard = calculate_zeta3_standard(error_bound_target)
print("--------------------------------------------------------------------------------")
calculated_zeta3_series2, ps_series2, eb_series2, acc_idx_series2 = calculate_zeta3_series2(error_bound_target)
print("--------------------------------------------------------------------------------")
calculated_zeta3_series3, ps_series3, eb_series3, acc_idx_series3 = calculate_zeta3_series3(error_bound_target)
print("--------------------------------------------------------------------------------")

print(f"Known Zeta(3) value: {known_zeta3:.15f}")
print(f"Difference (Standard): {abs(calculated_zeta3_standard - known_zeta3):.15f}")
print(f"Difference (Series 2): {abs(calculated_zeta3_series2 - known_zeta3):.15f}")
print(f"Difference (Series 3): {abs(calculated_zeta3_series3 - known_zeta3):.15f}")
```
